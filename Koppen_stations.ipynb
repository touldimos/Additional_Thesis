```python
import os
os.environ['PROJ_LIB'] = 'C:/Users/owner/Anaconda3/Lib/site-packages/mpl_toolkits/basemap'
from mpl_toolkits.basemap import Basemap

plt.figure(figsize = (20, 20))
axs = plt.subplot(111)
m=Basemap()
m.etopo(scale=0.5, alpha=0.5)
m.shadedrelief()
m.drawcoastlines(0.75)
parallels = np.arange(-90.,91.,15.)
m.drawparallels(parallels,labels=[False,True,True,False])
meridians = np.arange(-180.,181.,25.)
m.drawmeridians(meridians,labels=[True,False,False,True])

a = network[network['Koppen']==1]
b = network[network['Koppen']==2]
c = network[network['Koppen']==3]
d = network[network['Koppen']==4]
e = network[network['Koppen']==5]
# h = network[network['Koppen']==6]

axs.scatter(a['lon'], a['lat'], marker='o', s = 40, edgecolors = 'black', c = '#006600', zorder = 10, label = 'Class A - Tropical')
axs.scatter(b['lon'], b['lat'], marker='o', s = 40, edgecolors = 'black', c = '#ff0000', zorder = 10, label = 'Class B - Dry')
axs.scatter(c['lon'], c['lat'], marker='o', s = 40, edgecolors = 'black', c = '#ffff4d', zorder = 10, label = 'Class C - Temperate')
axs.scatter(d['lon'], d['lat'], marker='o', s = 40, edgecolors = 'black', c = '#00ff00', zorder = 10, label = 'Class D - Continental')
axs.scatter(e['lon'], e['lat'], marker='o', s = 40, edgecolors = 'black', c = '#1a1aff', zorder = 10, label = 'Class E - Polar')
# axs.scatter(h['lon'], h['lat'], marker='o', s = 40, edgecolors = 'black', c = '#ff33ff', zorder = 10, label = 'Class H - Highland')
axs.legend(loc = 'lower left')
plt.plot();
```
