```python
import pandas as pd
import numpy as np
    
def read_ecad_sum(path, VAL):
    ecad = pd.read_csv(path, names = ['STAID', 'SOUID', 'DATE', VAL, 'QUALITY'], skiprows = 21, parse_dates = ['DATE']).set_index('DATE')
    ecad = ecad[ecad['QUALITY'] == 0]
    output = pd.DataFrame()
    if ecad[VAL].count() >= 365 * 20 and ecad.index.max() >= pd.to_datetime('01/01/2000') and ecad.index.min() < pd.to_datetime('01/01/2000'):
        output[VAL] = ecad[VAL].resample(rule = 'M').sum()
    return output

def read_ecad_mean(path, VAL):
    ecad = pd.read_csv(path, names = ['STAID', 'SOUID', 'DATE', VAL, 'QUALITY'], skiprows = 21, parse_dates = ['DATE']).set_index('DATE')
    ecad = ecad[ecad[VAL] != -9999]
    if VAL == 'RS':
        ecad = ecad[ecad[VAL] <= 500]
    output = pd.DataFrame()
    if ecad[VAL].count() >= 365 * 20 and ecad.index.max() >= pd.to_datetime('01/01/2000') and ecad.index.min() < pd.to_datetime('01/01/2000'):
        output[VAL] = ecad[VAL].resample(rule = 'M').mean()
    return output
```

```python
U = read_ecad_mean(r"C:\Users\owner\Downloads\755\FG_STAID000755.txt", 'U2')/10
RH = read_ecad_mean(r"C:\Users\owner\Downloads\755\HU_STAID000755.txt", "RH")
Rs = read_ecad_mean(r"C:\Users\owner\Downloads\755\QQ_STAID000755.txt", "R")
P = read_ecad_sum(r"C:\Users\owner\Downloads\755\RR_STAID000755.txt", "P")
Tmean = read_ecad_mean(r"C:\Users\owner\Downloads\755\TG_STAID000755.txt", "T")/10
Tmax = read_ecad_mean(r"C:\Users\owner\Downloads\755\TN_STAID000755.txt", "Tmax")/10
Tmin = read_ecad_mean(r"C:\Users\owner\Downloads\755\TX_STAID000755.txt", "Tmin")/10
```

```python
dfs_array = [U, RH, Rs, P, Tmean, Tmax, Tmin]
data = pd.concat(dfs_array, axis=1).dropna()
data.to_excel('Monthly_755.xlsx')
```
