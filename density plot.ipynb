```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
import pandas as pd

data = pd.read_csv(r"C:\Users\owner\Google Drive 2\MSc\TU DELFT - Water Managment\Additional Thesis\Results\45(2).txt", delimiter = '\t')
# Generate fake data
x = data['x']
y = data['y']

# Calculate the point density
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)

# Sort the points by density, so that the densest points are plotted last
idx = z.argsort()
x, y, z = x[idx], y[idx], z[idx]

fig, ax = plt.subplots(figsize=(10, 8))

from matplotlib.colors import LinearSegmentedColormap

plot1 = ax.scatter(x, y, c = z, s = 50, zorder = 2)
ax.set_xscale('log')
ax.set_xlim(0.1, 300)
ax.tick_params(labelsize = 25)
ax.set_ylabel((u'$E_{r}$ (mm month $^{-1}$)'), fontsize  = 25)
ax.set_xlabel((u'Monthly $c_{th}$'), fontsize  = 25)
# ax.set_xlim(0, 70)
# ax.set_ylim(0, 400)
ax.axhline(45, label = u'$E_{r}$ = 45 mm month $^{-1}$', color = 'black')
plt.grid(zorder = 1)
plt.legend(fontsize  = 25)
cbar = plt.colorbar(plot1)
cbar.ax.tick_params(labelsize=25)
plt.show()
fig.savefig(r'C:\Users\owner\Google Drive 2\MSc\TU DELFT - Water Managment\Additional Thesis\Results\45.png', format='png', dpi=1200)
```
